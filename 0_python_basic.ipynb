### 1. 변수와 출력함수

### 3. 조건문 (if분기문, 다중if문)

```python
# if 중첩문
x=9
if x>0:
    if x<10:
        print('10보다 작은 자연수')

x=9
if x>0 and x<10:
    print('10보다 작은 자연수')

x=9
if 0<x<10:
    print('10보다 작은 자연수')

# if 분기문
x=10
if x>0:
    print('양수')
else:
    print('음수')

x=78
if x>=90:
    print('A')
elif x>=80:
    print('B')
elif x>=70:
    print('C')
else:
    print('F')
```

### 4. 반복문(for, while, break, continue)

```python
# 반목문 (for, while)
a = range(10) # 0 ~ 9까지
a = range(1,11) # 1 ~ 10까지
print(list(a))

for i in range(10):
    print(i)

for i in range(10,0,-1):  # -1씩 줄어든다
    print(i)

i=1
while i<=10:
    print(i)
    i+=1

i=10
while i>=1:
    print(i)
    i-=1

i=1
while True: #무한반복
    print(i)
    if i==10: # 10일때 반복문 멈추기
        break
    i+=1

for i in range(1,11):
    if i%2==0: # 짝수는 건너뛰고 홀수만 출력하고
        continue
    print(i)
```

### 5. 반복문을 이용한 문제 풀이

```python
# 1. 1부터 N까지 홀수 출력하기
n=int(input())
for i in range(1,n+1):
    if i%2==1:
        print(i)

#2. 1부터 N까지의 합 구하기
n=int(input())
sum=0
for i in range(1,n+1):
    sum+=i
print(sum)

#3. N의 약수 출력하기
n=int(input())
for i in range(1,n+1):
    if n%i==0:
        print(i, end=' ')
```

### 6. 중첩반복문 (2중 for문)

```python
for i in range(5):
    for j in range(i+1):
        print('*', end=' ')
    print()

for i in range(5):
    for j in range(5-i):
        print('*', end=' ')
    print()
```

### 7. 문자열과 내장함수

```python
msg ='It is Time'
print(msg.upper()) #대문자
print(msg.lower()) #소문자
tmp = msg.upper()
print(tmp.find('T')) # 대문자 T를 찾아라
print(tmp.count('T')) # 대문자 T가 몇개 있는가
print(msg[:2]) # 0~1번까지 슬라이싱
print(len(msg)) # msg의 총 길이

for i in range(len(msg)):
    print(msg[i], end=' ')

for x in msg:
    print(x, end=' ')

#대문자만 뽑는 경우
for x in msg:
    if x.isupper():
        print(x, end=' ')

#공백을 빼고 문자만 출력하는 경우
for x in msg:
    if x.isalpha():
        print(x, end='')

tmp='AZ'
for x in tmp:
    print(ord(x))  #ord : 아스키코드를 출력한다. (A는 65, a는 97)

tmp=65
print(chr(tmp))   #chr : 문자를 출력한다.
```

### 8. 리스트와 내장함수(1)

```python
import random as r
a=[]
b=list()

a=[1,2,3,4,5]
print(a)

b=list(range(1,11)) #1부터 10까지 리스트화
print(b)

c=a+b  # 리스트 합치기
print(c)

a.append(6)   # 추가
a.insert(3,7) # 3번 인덱스에 값 7을 넣어라
print(a)

a.pop()   # 맨 뒤 리스트 제거
print(a)
a.pop(3)  # 3번 인덱스 값 빼라
print(a)

a.remove(4)  # 4라는 값을 제거해라
print(a)

print(a.index(5)) # 5라는 값의 인덱스를 출력해라

a=list(range(1,11))
print(a)
print(sum(a))
print(max(a))
print(min(a))
print(min(7,3,5)) #인자값들 중에서 가장 작은 값 출력

r.shuffle(a) # 무작위 셔플
print(a)
a.sort() #오름차순 정렬
print(a)
a.sort(reverse=True) # 내림차순
print(a)
a.clear() #초기화
print(a)
```

### 9. 리스트와 내장함수(2)

```python
a=[23,12,36,53,19]
print(a[:3])
print(a[1:4])
print(len(a))

for x in enumerate(a): # enumerate : 인덱스와 값을 같이 출력
    print(x)

b=(1,2,3,4,5) # tuple
print(b[0])
# b[0]=7 # 리스트는 값 변경이 가능하지만 튜플은 값 변경이 불가하다!

for x in enumerate(a):
    print(x[0], x[1])

for index, value in enumerate(a):
    print(index, value)

#all
if all(60>x for x in a): # 하나라도 거짓이면 거짓 리턴!
    print("YES")
else:
    print("NO")

#any
if any(15>x for x in a): # 한개만 참이면 참 리턴!
    print("YES")
else:
    print("NO")
```

### 10. 2차원 리스트 생성과 접근

```python
a=[[0]*3 for _ in range(3)]
print(a)
a[0][1]=1
print(a)

for x in a:
    print(x)

for x in a:
    for y in x:
        print(y, end=' ')
    print()
```

### 11. 함수 만들기

```python
# #방식1
# def add(a, b):
#     c=a+b
#     print(c)

# #방식2
# def ad(a,b):
#     c=a+b
#     return c


# add(3,2)

def add(a,b):
    c=a+b
    d=a-b
    return c,d

print(add(3,2))

# 소수만 출력하기
def isPrime(x):
    for i in range(2,x):
        if x%i==0:
            return False
    return True

a=[12,13,7,9,19]
for x in a:
    if isPrime(x):
        print(x, end=' ')
```

### 12. 람다함수

```python
def plus_one(x):
    return x+1

print(plus_one(1))

# --> 람다함수로 표현하기
plus_two = lambda x:x+2
print(plus_two(1))

a=[1,2,3]
print(list(map(plus_one, a)))
print(list(map(lambda x:x+1, a)))
# sort를 사용할 때 람다식을 자주 쓴다
```
